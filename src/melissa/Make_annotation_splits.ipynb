### Make annotation splits

```python
import pandas as pd
import json
from datetime import datetime
```

```python
start_data = '../../data/five_models_normalized_categorical_2023-04-17_23.29.43.888250.json'
df = pd.read_json(start_data)
print(df.columns)
df.head(3)
```

```python
our_samples_df = df.loc[df.annotator.isna()].sample(n=700, random_state=42)  
assert our_samples_df.shape == (700, df.shape[1])

shared_samples_df = our_samples_df.iloc[:100]
assert shared_samples_df.shape[0] == 100

mel_samples_df = our_samples_df.iloc[100:300]
mel_samples_df = pd.concat([shared_samples_df, mel_samples_df])
print(mel_samples_df.shape)

sarik_samples_df = our_samples_df.iloc[300:500]
sarik_samples_df = pd.concat([shared_samples_df, sarik_samples_df])
print(sarik_samples_df.shape)

anu_samples_df = our_samples_df.iloc[500:700]
anu_samples_df = pd.concat([shared_samples_df, anu_samples_df])
print(anu_samples_df.shape)

timestamp = str(datetime.now()).split('.')[0].replace(' ', '.').replace(':', '').replace('-', '')
print(timestamp)
#print(str(timestamp).split('.')[0].replace(' ', '.').replace(':', '').replace('-', ''))

#print(f'mel_samples_{timestamp}.tsv')
mel_samples_df.to_csv(f'mel_samples_{timestamp}.tsv', sep='\t')
sarik_samples_df.to_csv(f'sarik_samples_{timestamp}.tsv', sep='\t')
anu_samples_df.to_csv(f'anu_samples_{timestamp}.tsv', sep='\t')
```
