```python
textblob_df = pd.read_csv(polarity_files[0], sep='\t')
textblob_df.head()
```

```python
vader_df = pd.read_csv(polarity_files[1], sep='\t')
vader_df.head()
```

# Combine datasets

```python
polarity_df = textblob_df.merge(vader_df, how='inner', on='Unnamed: 0')
polarity_df.head()
```

```python
orig_cols = polarity_df.columns
print(f'original columns:{orig_cols}')
new_cols = [str(x).replace('_x', '_textblob').replace('_y', '_vader') for x in polarity_df.columns]
print(f'\nNew columns: {new_cols}')
```

```python
polarity_df.columns = new_cols
polarity_df.head()
```

# Quick polarity score inspection

A first look at vader vs textblob polarization scores shows that they don't really agree as well as one might expect.  The first and last few rows show the two models have relatively different scores.

```python
polarity_df['textblob_vader_diff'] = abs(polarity_df['Polarity_textblob'] - polarity_df['Polarity_vader'])
polarity_df[['Polarity_textblob', 'Polarity_vader', 'textblob_vader_diff']]  # these are pretty wildly different...
```

## Textblob and Vader consistency
The average polarity score for textblob is 0.05, and for vader is 0.01.  Looking at the average _difference_ in the two scores, we see an average difference of 0.04 (not too bad, really).  But there are a number of tweets that have a large difference, as shown by the **standard deviation of 0.40**.

Vader seems to skew negative at the first quartile mark as compared to textblob.

```python
polarity_df[['Polarity_textblob', 'Polarity_vader', 'textblob_vader_diff']].describe()
```

## Observe distance between two scores

```python
# histogram of textblob-vader differences
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

figure(figsize=(12, 6), dpi=80)

polarity_df['textblob_vader_diff'].hist(bins=25)
plt.title('Histogram of distance between two scores\n(score range is -1 to 1)')
plt.xlabel('Absolute Distance')
plt.ylabel('Count of tweets')
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

figure(figsize=(12, 6), dpi=80)

# Make ordered datasets
vader_desc = polarity_df['Polarity_vader'].sort_values(ascending=False)
#print(vader_desc[:10])
#type(vader_desc)  # series
#vader_desc.index[:10]

# Order textblob in same order as vader
textblob_desc = polarity_df['Polarity_textblob'].loc[vader_desc.index]
print(textblob_desc.index[:10])
print(textblob_desc.values[:10])

# Plot the two series
print(vader_desc.values[:10])
print(textblob_desc.values[:10])
x = range(vader_desc.index.max()+1)
plt.plot(x, textblob_desc.values, label='textblob corresponding polarity')
plt.plot(x, vader_desc.values, label='vader polarity, descending')
plt.title('Sorting Tweets by Vader Polarity, observe Textblob Polarity')
plt.xlabel('Tweets in vader-descending order')
plt.ylabel('Polarity (-1 is most negative, +1 most positive)')
plt.legend()
plt.show()
```
