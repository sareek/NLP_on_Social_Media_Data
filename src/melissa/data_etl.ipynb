# Merge data

```python
# Gather input files
json_files = !ls ../../data/*.json
print(json_files)

# Get columns
tmp = pd.read_json(json_files[0])
columns = list(tmp.columns)
columns.append('keyword')

# Combine all datasets
df = pd.DataFrame(columns=columns)
for f in json_files:
    tmp = pd.read_json(f)
    tmp['keyword'] = f.split('data/')[1].split('_project_tweets')[0]
    df = df.append(tmp)
    print(df.shape)

# Inspect data
df.reset_index(inplace=True, drop=True)
display(df.head())
display(df.tail())

# Save combinied dataset
timestamp = str(datetime.now()).replace(' ', '_').replace(':', '.')
print(timestamp)

outpath = os.path.join('..', '..', 'data', f'project_data.{timestamp}.json')
df.to_json(outpath)
print(f'Wrote full dataset to {outpath}')

# csv version:
csv_outpath = os.path.join('..', '..', 'data', f'project_data.{timestamp}.csv')
df.to_csv(csv_outpath, )
print(f'Wrote full dataset to {csv_outpath}')
```
